```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from time import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codeheroku/Introduction-to-Machine-Learning/master/Face%20Recognition%20Using%20PCA/face_data.csv")
df.head()
```

```python
x = df.iloc[:,:4096]
y = df.iloc[:,4096:]
```

```python
print("Dimension of each image:",np.array(x.iloc[0]).reshape(1,4096).shape)
print("Dimension of dataset",x.shape)
```

```python
print(y.target.unique())
print("No of classes in dataset:",len(y.target.unique()))
```

```python
#Sample image
plt.imshow(np.array(x.iloc[2]).reshape(64, 64), cmap='gray')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 23)
```

```python
x_train.head()
```

```python
x_train.T.shape
```

```python
avg_face = []

for index,col in x_train.T.iterrows():
    avg_face.append(np.mean(col))
```

```python
avg_face = np.array(avg_face)
```

```python
avg_face = avg_face.reshape(1,4096)
```

```python
avg_face.shape
```

```python
x_train = x_train.sub(avg_face.squeeze(), axis=1)
```

```python
x_train.head()
```

```python
x_train.reset_index(inplace=True)
```

```python
x_train.drop("index",axis=1,inplace=True)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
y_train.head()
```

```python
y_train.reset_index(inplace=True)
```

```python
y_train.head()
```

```python
pca = PCA(n_components=200).fit(x_train)
```

```python
pca
```

```python
def show_eigenfaces(pca):

    fig, axes = plt.subplots(4, 8, figsize=(12, 8),subplot_kw={'xticks':[], 'yticks':[]})
    for i, ax in enumerate(axes.flat):
        ax.imshow(pca.components_[i].reshape(64, 64), cmap='gray')
        ax.set_title("PC " + str(i+1))
    plt.show()
```

```python
x_train.head()
```

```python
Xtrain_pca = pca.transform(x_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
show_eigenfaces(pca)
```

```python
Xtrain_pca.shape
```

```python
x_test = x_test.sub(avg_face.squeeze(), axis=1)
```

```python
x_test.reset_index(inplace=True)
```

```python
x_test.head()
```

```python
x_test.drop("index",axis=1,inplace=True)
```

```python
Xtest_pca = pca.transform(x_test)
```

```python
Xtest_pca.shape
```

```python
import sys
print(sys.maxsize)
```

```python
result = []
for item in Xtest_pca:
    min_dist = sys.maxsize
    index = -1 
    for i,sample in enumerate(Xtrain_pca):
        dist = abs(sample-item).sum()
        if dist<min_dist:
            min_dist = dist
            index = i
    result.append(y_train.target[index])
```

```python
result
```

```python
y_test = y_test.reset_index().drop("index",axis=1)
```

```python
y_test
```

```python
y_test_match = list(y_test.target)
```

```python
correct = 0
incorrect = []
for index,item in enumerate(y_test_match):
    if item==result[index]:
        correct+=1
    else:
        incorrect.append(index)
print(correct)
```

```python
incorrect
```

```python
print("accuracy:",correct/len(y_test))
```
