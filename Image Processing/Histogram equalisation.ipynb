```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('flower1.jpg')
```

```python
plt.imshow(image,cmap='gray')
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
plt.imshow(gray,cmap='gray')
```

```python
hist = np.zeros(256)

for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        hist[gray[i][j]] = hist[gray[i][j]]+1
```

```python
hist
```

```python
index = np.arange(len(hist))
plot1 = plt.bar(index,hist)
```

```python
pdf = [np.around(i/(gray.shape[0]*gray.shape[1]),decimals=5) for i in hist];pdf
```

```python
cdf = []
cdf.append(pdf[0])

for i in range(1,len(pdf)):
    cdf.append(np.around(cdf[i-1]+pdf[i],decimals=6))
    
cdf
```

```python
eq_levels = np.around(np.multiply(cdf,255))
```

```python
eq_levels
```

```python
eq_hist = np.zeros(eq_hist.shape)

for i in range(len(eq_hist)):
    eq_hist[i] = eq_levels[i]
    
eq_hist = np.array(eq_hist,dtype=np.uint8)
```

```python
eq_hist
```

```python
final_hist = np.zeros(256)

for i in range(len(final_hist)):
    final_hist[eq_hist[i]]+=hist[i]
    
final_hist
```

```python
plt.bar(index,final_hist)
```
