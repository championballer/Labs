Grey and 3 channel separation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("flower.jpeg", cv2.IMREAD_COLOR)
```

```python
image[:, :, 0], image[:, :, 2] = np.array(image[:, :, 2]), np.array(image[:, :, 0])
```

```python
shape = image.shape
gray_image = np.zeros(shape[:2])
```

```python
for i in range(shape[0]):
    for j in range(shape[1]):
            red   = image[i][j][0]
            green = image[i][j][1]
            blue  = image[i][j][2]
            gray_image[i][j] = 0.3 * red + 0.59 * green + 0.11 * blue
```

```python
plt.imshow(gray_image, cmap='gray')
```

```python
plt.imshow(image)
```

```python
red_channel = np.array(image)
red_channel[:, :, 1:3] = 0
plt.imshow(red_channel)
```

```python
green_channel = np.array(image)
green_channel[:, :, 0] = 0
green_channel[:, :, 2] = 0
plt.imshow(green_channel)
```

```python
blue_channel = np.array(image)
blue_channel[:, :, 0:2] = 0
plt.imshow(blue_channel)
```
