```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cars.jpg')
plt.imshow(img)
plt.title('Traffic Picture 1')
plt.show()

shape = img.shape
gray_image = np.zeros(shape[:2])
for i in range(shape[0]):
    for j in range(shape[1]):
            red   = img[i][j][0]
            green = img[i][j][1]
            blue  = img[i][j][2]
            gray_image[i][j] = 0.3 * red + 0.59 * green + 0.11 * blue
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
print(img.shape)
print(gray_image.shape)
```

```python
def pad(img,shp):
    p=np.zeros((shp[0]+2,shp[1]+2))
    p[1:-1,1:-1]=np.copy(img)
    p[0,1:-1],p[-1,1:-1]=img[0],img[-1]
    p[1:-1,0],p[1:-1,-1]=img[:,0],img[:,-1]
    p[0,0],p[0,-1]=img[0,0],img[0,-1]
    p[-1,0],p[-1,-1]=img[-1,0],img[-1,-1]
    return p
```

```python
def sobel_filter_trial(img):
    sabel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    sabel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    shp = img.shape
    shpm=(3,3)
    padded_img=pad(img,shp)
    grad_list=[]
    out=np.zeros(shp)
    for i in range(shp[0]):
        for j in range(shp[1]):
            g_x=np.multiply(padded_img[i:i+shpm[0],j:j+shpm[1]],sabel_x).sum()
            g_y=np.multiply(padded_img[i:i+shpm[0],j:j+shpm[1]],sabel_y).sum()
            if g_y!=0 or g_x!=0:
                out[i,j]=0
                if g_x==0:
                    rad=np.arctan(float("inf"))
                else:
                    rad=np.arctan(g_y/g_x)
                grad_list.append([i,j,rad])
                
            else:
                out[i,j]=255
    out=np.array(out, dtype = np.uint8)
    plt.imshow(out,cmap='gray')
    plt.show()
```

```python
def sobel_filter(img):
    sabel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    sabel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    shp = img.shape
    shpm=(3,3)
    padded_img=pad(img,shp)
    grad_list=[]
    out=np.zeros(shp)
    for i in range(shp[0]):
        for j in range(shp[1]):
            g_x=np.multiply(padded_img[i:i+shpm[0],j:j+shpm[1]],sabel_x).sum()
            g_y=np.multiply(padded_img[i:i+shpm[0],j:j+shpm[1]],sabel_y).sum()
            if g_y!=0 or g_x!=0:
                out[i,j]=np.sqrt(np.square(g_x)+np.square(g_y))
                if g_x==0:
                    rad=np.arctan(float("inf"))
                    print("x:",i,"y:",j,"angle:",rad)
                    out
                else:
                    rad=np.arctan(g_y/g_x)
                    print("x:",i,"y:",j,"angle:",rad)
                grad_list.append([i,j,rad])
                
            else:
                out[i,j]=0
    axes = plt.gca()
    axes.set_xlim([0,img.shape[1]])
    axes.set_ylim([img.shape[0],0])
    out=np.array(out, dtype = np.uint8)
    plt.imshow(out,cmap='gray')
    plt.show()
    cv2.imwrite('01.png',out)
```
